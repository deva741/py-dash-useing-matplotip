```python
import sqlite3

# Connect to SQLite database (it will create 'example.db' if it doesn't exist)
conn = sqlite3.connect('example.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create a table (if not exists)
cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER NOT NULL
)
''')

# Insert data into the table
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Alice", 30))
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Bob", 25))
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Mouli", 30))
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Keerthy", 25))

# Commit the changes
conn.commit()

# Retrieve and print all data
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
import queue

# Create a FIFO queue ( First IN - First OUT)
q = queue.Queue()

# Enqueue elements
q.put(10)
q.put(20)

# Dequeue elements from the queue
print(q.get())  # Output: 10
print(q.get())  # Output: 20

# Check if the queue is empty
print(q.empty())  # Output: True

# Get the size of the queue
print(q.qsize())  # Output: 0
```

```python
from collections import deque

# Create a deque object
q = deque()

# Enqueue elements into the queue
q.append(10)
q.append(20)
q.append(30)

# Dequeue elements from the queue
print(q.popleft())  # Output: 10
print(q.popleft())  # Output: 20
print(q.popleft())  # Output: 30

# Check if the queue is empty 
print(len(q) == 0)   # Output: True
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate sample data using NumPy
np.random.seed(0)
days = np.arange(1, 11)  # Days 1 to 10
sales = np.random.randint(100, 500, size=10)  # Random sales data

# Step 2: Create a DataFrame using Pandas
df = pd.DataFrame({
    'Day': days,
    'Sales': sales
})
print("Sales Data:\n", df)
```

```python
# Step 3: Plot the data using Matplotlib
plt.figure(figsize=(8, 4))
plt.plot(df['Day'], df['Sales'], marker= 'o', linestyle='-', color='blue')
plt.title('Sales Over 10 Days')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load CSV using pandas
df = pd.read_csv(r'C:\Users\anand\Downloads\sales_data.csv')

# Step 2: Convert columns to NumPy arrays
days = df['Day'].to_numpy()
sales_a = df['Product_A_Sales'].to_numpy()
sales_b = df['Product_B_Sales'].to_numpy()

# Step 3: Display data
print("First 5 rows:\n", df.head())
```

```python
# Step 4: Plot the sales data
plt.figure(figsize=(8, 5))
plt.plot(days, sales_a, label='Product A', marker='o', color='blue')
plt.plot(days, sales_b, label='Product B', marker='s', color='Purple')
plt.title('Sales Comparison: Product A vs Product B')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
#1. Bar Chart
bar_width = 0.35
x = np.arange(len(days))

plt.bar(x, sales_a, width=bar_width, label='Product A', color='skyblue')
plt.bar(x + bar_width, sales_b, width=bar_width, label='Product B', color='lightgreen')

plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Bar Chart: Product A vs B Sales')
plt.xticks(x + bar_width / 2, days)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
#2. Stacked Area Chart
plt.stackplot(days, sales_a, sales_b, labels=['Product A', 'Product B'], colors=['blue', 'green'])
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Stacked Area Chart: Product A and B Sales')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
#3. Scatter Plot
plt.scatter(days, sales_a, color='blue', label='Product A', marker='o')
plt.scatter(days, sales_b, color='green', label='Product B', marker='s')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Scatter Plot: Sales of Products A & B')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
#4. Subplot: Line Charts Side by Side
# Product A
axs[0].plot(days, sales_a, marker='o', color='blue')
axs[0].set_title('Product A Sales')
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Sales')
axs[0].grid(True)

# Product B
axs[1].plot(days, sales_b, marker='s', color='green')
axs[1].set_title('Product B Sales')
axs[1].set_xlabel('Day')
axs[1].set_ylabel('Sales')
axs[1].grid(True)

plt.tight_layout()
plt.show()

#You can try all of these by replacing your plt.plot(...) section with any of the above.
```
